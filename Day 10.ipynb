```python
#!/bin/python3

import math
import os
import random
import re
import sys



if __name__ == '__main__':
    n = int(input().strip())
    x = []
    while n>0:
        if n%2==0:
            x.append(0)
        else:
            x.append(1)
        n = n//2   
    x.reverse()

    i = 0
    new = []
    while(i<len(x)):
        count = 1

        while True:
            if i<len(x)-1 and x[i]==1 and x[i+1]==1:
                count+=1
            else:
                break
            i+=1
        new.append(count)
        i+=1
    print(max(new))
```
