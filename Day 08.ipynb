```python
# Enter your code here. Read input from STDIN. Print output to STDOUT
n = int(input())
Dict = {}

for i in range(n):
    x = input()
    Dict[x.split(" ")[0]]= x.split(" ")[1]
    
try:
    inp = input()
    while inp!= "":
        if inp in Dict:
            print(inp+"="+Dict[inp])
        else:
            print("Not found")
        inp = input()
except:
    pass

        
```
