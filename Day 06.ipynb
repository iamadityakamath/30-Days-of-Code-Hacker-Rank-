```python
# Enter your code here. Read input from STDIN. Print output to STDOUT
T = int(input())

for j in range(T):
    x = input()
    even = ''
    odd = ''
    if len(x)>=2:
        for i in range(len(x)):
            if i%2==0:
                even = even+x[i]
            else:
                odd = odd+x[i]
        print(even + " "+odd)

        
```
